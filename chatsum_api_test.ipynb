````python
import json
import codecs, random
dialog_list = json.loads(codecs.open("data/NaturalConv_Release_20210318/dialog_release.json", "r", "utf-8").read())
random.shuffle(dialog_list)
print(dialog_list[0:2])
````

````python
all_chats = []
for dialog in dialog_list:
    for index, chat in enumerate(dialog['content']):
        person = 0 if index % 2 == 0 else 1
        all_chats.append(dialog['dialog_id']+"_"+str(person)+"："+chat)
print(all_chats[:10])

user_msgs=[]
chat_group_len=60
chat_group_num=len(all_chats)//chat_group_len
for i in range(10):
    chat_group = all_chats[i*chat_group_len:(i+1)*chat_group_len]
    # random.shuffle(chat_group)
    user_msg="".join(["<|im_start|>{}<|im_end|>\n".format(chat) for chat in chat_group])
    user_msgs.append((user_msg, ""))
print(user_msgs[0])

system_msg = '''你是一个优秀的对话聊天助手。基于所有多轮聊天记录，你需要逐步完成如下任务，首先，将聊天记录归纳出最多5个主要的话题，话题名称尽量简洁；然后针对每个话题给出总结和要点，尽量给出每个人的观点或态度。提示：`<|im_start|>`代表每条聊天的开始，`<|im_end|>\n`代表每条聊天的结束，聊天记录均在"""内。'''
````

````python
def get_response_of_openai(user_msg, system_msg="", model_engine="gpt-35-turbo", max_tokens=1500):
    openai.api_type = "azure"
    openai.api_base = "https://agi-v02.openai.azure.com/"
    openai.api_version = "2023-07-01-preview"
    openai.api_key = os.getenv("OPENAI_API_KEY")

    response = openai.ChatCompletion.create(
      engine=model_engine,
      messages=[{"role": "system", "content": system_msg},
                  {"role": "user", "content": user_msg}],
      temperature=0.0,
      max_tokens=max_tokens,
      top_p=0.95,
      frequency_penalty=0,
      presence_penalty=0,
      stop=None)
    
    message = response.choices[0]["message"]
    if "content" in message:
        return message["content"]
    else:
        filtered_keys = [key for key, value in response.choices[0]["content_filter_results"].items() if value["filtered"]]
        # print(response)
        return "filtered! reasons: {}".format(str(filtered_keys))
    
    # print('engine: %s' % model_engine)


def get_response_of_spark(user_msg, system_msg="", model_engine="generalv2", max_tokens=1500):
    #以下密钥信息从控制台获取
    appid = "b65537a8"     #填写控制台中获取的 APPID 信息
    api_secret = "xxx"   #填写控制台中获取的 APISecret 信息
    api_key ="xxx"    #填写控制台中获取的 APIKey 信息

    #用于配置大模型版本，默认“general/generalv2”
    # domain = "general"   # v1.5版本
    # domain = "generalv2"    # v2.0版本
    #云端环境的服务地址
    # Spark_url = "ws://spark-api.xf-yun.com/v1.1/chat"  # v1.5环境的地址
    Spark_url = "ws://spark-api.xf-yun.com/v2.1/chat"  # v2.0环境的地址

    def getText(role,content): 
        text=[]
        jsoncon = {}
        jsoncon["role"] = role
        jsoncon["content"] = content
        text.append(jsoncon)
        return text

    msg = "{}，聊天记录均在```内。```{}```".format(system_msg, user_msg)
    question = getText("user",msg)

    import SparkApi
    SparkApi.answer =""
    SparkApi.main(appid,api_key,api_secret,Spark_url,model_engine,question)
    # getText("assistant",SparkApi.answer)
    return SparkApi.answer

def get_response_of_minimax(user_msg, system_msg="", model_engine="abab5.5-chat", max_tokens=1034):
    import requests

    group_id="xxx"
    api_key="xxx"

    url = "https://api.minimax.chat/v1/text/chatcompletion_pro?GroupId=" + group_id

    payload = {
        "bot_setting": [
            {
                "bot_name": "ML Assistant",
                "content": system_msg,
            }
        ],
        "messages": [{"sender_type": "USER", "sender_name": "user", "text": ""}],
        "reply_constraints": {"sender_type": "BOT", "sender_name": "ML Assistant"},
        "model": model_engine,
        "tokens_to_generate": max_tokens,
        "temperature": 0.01,
        "top_p": 0.95,
    }
    headers = {"Content-Type": "application/json", "Authorization": "Bearer " + api_key}

    payload['messages'][0]['text'] = user_msg
    response = requests.request("POST", url, headers=headers, json=payload)

    # print(response.status_code)
    return print(response.json()['reply'])
````

````python
def generate_response(user_msg, system_msg, model_engine, max_tokens=1500):
    """
    Generate a response using the specified model engine and parameters.
    """
    model_comp = {
        "gpt-4":"openai", "gpt-35-turbo-16k":"openai", "gpt-35-turbo":"openai",
        "general":"spark", "generalv2":"spark",
        "abab5.5-chat":"minimax"
    }
    comp=model_comp[model_engine]
    if comp == "openai":
        response = get_response_of_openai(user_msg, system_msg=system_msg, model_engine=model_engine, max_tokens=max_tokens)
    elif comp == "claude":
        pass
    elif comp == "spark":
        response = get_response_of_spark(user_msg, system_msg=system_msg, model_engine=model_engine, max_tokens=max_tokens)
    elif comp == "minimax":
        response = get_response_of_minimax(user_msg, system_msg=system_msg, model_engine=model_engine, max_tokens=max_tokens)
    elif comp == "qwen":
        pass
    else:
        raise ValueError("Invalid model engine specified.")
    return response
````
